```python
import matplotlib as matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import pandas as pd, seaborn as sns
sns.set_style('white')
import matplotlib.pyplot as plt

config = {
  'toImageButtonOptions': {
    'format': 'png', # one of png, svg, jpeg, webp
    'filename': 'newplot',
    'height': 500,
    'width': 900,
    'scale': 5 # Multiply title/legend/axis/canvas sizes by this factor
  }
}

df = pd.read_csv('tum_living.csv', quotechar=",")
df.head()
```

```python
df.describe()
```

```python
ax = sns.lineplot(x="Rooms", y="Rent Price", data=df, errorbar=None)
ax.set(xlabel='Rooms', ylabel='Rent Price ()')
plt.title('Average Rent Price per Total Rooms according to TUM Living 2022', fontsize = 12)
```

```python
ax = sns.lineplot(x = "Area", y = "Rent Price", data=df, errorbar=None)
ax.set(xlabel='Area (m)', ylabel='Rent Price ()')
plt.title('Rent Price per Total Area according to TUM Living 2022', fontsize = 12)
```

```python
ax = sns.stripplot(data=df, x="Rooms", y="Rent Price")
ax.set(xlabel='Rooms', ylabel='Rent Price ()')
plt.title('Rent Price per Total Rooms according to TUM Living 2022', fontsize = 12)
```
